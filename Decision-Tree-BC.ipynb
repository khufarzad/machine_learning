```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./dataset/Breast-Cancer.csv')
df.head()
# 1: True
# 2: False
```

```python
df = df.drop(labels=['id'], axis=1)
df.head()
```

```python
x = df.drop(labels='diagnosis', axis=1)
y = df.diagnosis
```

```python
from scipy.stats import zscore

scale = x.apply(zscore)  # scaled_x
scale.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=500)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(criterion='entropy', max_depth=6)
model
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_test[:10])
print(y_pred[:10])
```

```python
from sklearn import metrics

confusion = metrics.confusion_matrix(y_test, y_pred, labels=['M', 'B'])
confusion_df = pd.DataFrame(confusion,
                           index=['Real M', 'Real B'],
                           columns=['Predict M', 'Predict B'])
sns.heatmap(confusion_df, annot=True)
plt.show()
```
